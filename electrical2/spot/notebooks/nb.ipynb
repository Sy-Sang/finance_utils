```python
from finance_utils.electrical2.spot.probabilistic.curve import ForecastCurve
from finance_utils.electrical2.spot.probabilistic.market import ForecastMarket

from data_utils.stochastic_utils.distributions.basic_distributions import NormalDistribution,WeibullDistribution

import numpy
```

```python
real_dp = ForecastCurve([
    NormalDistribution(200,40),
    NormalDistribution(250,40),
    NormalDistribution(230,40),
    NormalDistribution(210,40),
], domain_min=0, domain_max=300)

real_rp = ForecastCurve([
    NormalDistribution(200 * 1.1,40 * 2),
    NormalDistribution(250 * 0.9,40 * 2),
    NormalDistribution(230 * 1, 40 * 1.5),
    NormalDistribution(210 * 0.9, 40 * 1.2),
], domain_min=0, domain_max=300)

real_q = ForecastCurve([
    WeibullDistribution(2, 5, 10),
    WeibullDistribution(1, 3, 10),
    WeibullDistribution(3, 10, 10),
    WeibullDistribution(2, 6, 10),
], domain_min=0, domain_max=30)

real_market = ForecastMarket(real_dp, real_rp, real_q)

forecast_dp = ForecastCurve([
    NormalDistribution(200 * 0.8, 40 * 0.5),
    NormalDistribution(250 * 1.2, 40 * 0.6),
    NormalDistribution(230 * 1.1, 40 * 0.5),
    NormalDistribution(210 * 1.2, 40 * 0.6),
], domain_min=0, domain_max=300)

forecast_rp = ForecastCurve([
    NormalDistribution(200 * 1.1 * numpy.random.uniform(0.6, 1.5), 40 * 2 * numpy.random.uniform(0,0.6)),
    NormalDistribution(250 * 0.9 * numpy.random.uniform(0.6, 1.5), 40 * 2 * numpy.random.uniform(0,0.6)),
    NormalDistribution(230 * 1 * numpy.random.uniform(0.6, 1.5), 40 * 1.5 * numpy.random.uniform(0,0.6)),
    NormalDistribution(210 * 0.9 * numpy.random.uniform(0.6, 1.5), 40 * 1.2 * numpy.random.uniform(0,0.6)),
], domain_min=0, domain_max=300)

forecast_q = ForecastCurve([
    WeibullDistribution(2, 5, 10),
    WeibullDistribution(1, 3, 10),
    WeibullDistribution(3, 10, 10),
    WeibullDistribution(2, 6, 10),
], domain_min=0, domain_max=30)

forecast_market = ForecastMarket(forecast_dp, forecast_rp, forecast_q)
```

```python
print(forecast_market.self_related_random_sample([0.9] * 4, 100).to_array().tolist())
```
